## **1. Introducción**

Los modelos de secuencias (en inglés sequence models) son las técnicas utilizadas cuando el orden y la secuencia de los datos aportan mucho valor predictivo.

**¿A qué nos referimos con secuencia?**

Una secuencia es una serie de elementos que se suceden unos a otros y guardan relación entre sí.

Voy a transcribirlo a un ejemplo fácil de asimilar. Cuando alguien nos pregunta el abecedario sabemos responder rápidamente sin dudarlo. Sin embargo, cuando tenemos que empezar el abecedario por una letra aleatoria nos cuesta un poco más, y si tenemos que decirlo de atrás para delante ya ni te cuento. Esto se debe a que tenemos metido en la cabeza el abecedario como una secuencia de letras y lo tenemos memorizado así. Esto es porque tanto las letras que preceden como las posteriores nos aportan mucha información para saber cuál toca.



Y, **¿Por qué no podemos utilizar modelos tradicionales o redes neuronales normales?**

Las redes neuronales estándar (NN a partir de ahora) presentan una serie de debilidades a la hora de afrontar este tipo de problemas:


*   Las NN están diseñadas para tener siempre el mismo tamaño de input y de output siempre. Sin embargo, en los problemas secuenciales, los inputs y los outputs pueden tener tamaños muy diferentes dependiendo de la observación. Por ejemplo, a la hora de traducir, cuando queremos traducir tres palabras tenemos un input y output mucho más pequeños que cuando queremos traducir un párrafo.
*  Las NN no comparten características entre las diferentes posiciones. Es decir, las NN asumen que cada input (y output) es independiente uno del otro, es decir, **no tienen ningún tipo de memoria**.

### ¿Y para qué se usan las RNN?

Como hemos comentado, son muy buenas sobretodo cuando nuestros datos son secuenciales:

* Predicción de acciones en bolsa

 * Por ejemplo, cuando el movil te sugiere palabras. Se fija en la ultima palabra que has escrito, y en las letras que estás escribiendo en ese momento para sugerirte las próximas letras o incluso palabras

Resulta que las RNN son una cadena de módulos repetidos, en los que cada módulo puede ser tan simple como una tanh de función de activación. Las LSTM tienen esta estructura de cadena de módulos, pero esta vez el módulo es un poco más complejo. Cada módulo, en vez de tener una simple función de activación, existen 4 neuronas que interactúan entre ellas.

Vamos a tratar de entenderlo, pero primero, os explico la nomenclatura:

<img src="https://image.ibb.co/na6W7d/rnn_arch_nomenclature.png" alt="rnn_arch_nomenclature" border="0" height="100">

En los diagramas de arriba, por cada línea viaja un vector, desde la salida de un nodo hasta las entradas de otros. Los círculos rosas indican operaciones elemento a elemento, como sumas de vectores, mientras que las cajas amarillas son capas neuronales que se aprenden al entrenar. Las líneas que se unen indican concatenar, y las que se separan, que el mismo contenido de la linea viaja a dos destinos distintos.

### La idea clave de las LSTMs

La clave es el estado de la celda, que está indicado en el diagrama como la línea que viaja por la parte de arriba:

<img src="https://image.ibb.co/hvJYLy/rnn_arch_cell_status.png" alt="rnn_arch_cell_status" border="0">

El estado de la celda es como una especie de cinta transportadora que viaja a lo largo de toda la arquitectura de la red con muy pocas interacciones (y las pocaas que tiene son lineales): lo cual implica que la información simplemente fluye sin ser modificada.

La parte ingeniosa es que las capas de la LSTM pueden (o no) aportar información a esta cinta transportadora, y esa decisión la toman las "puertas":

<img src="https://image.ibb.co/kjHSfy/rnn_arch_gate.png" alt="rnn_arch_gate" border="0">

Las puertas no son otra cosa que una forma de regular cuidadosamente la información que llega a la cinta transportadora. Están compuestas de una red neuronal con una activación de tipo sigmoide y una multiplicación elemento a elemento.

Así, la capa sigmoide da como salida un número entre 0 y uno, que implica cómo de importante es esa información para dejarla pasr a la cinta transportadora. Un 0 significa que no me importa, y un uno significa que es muy importante.

Como podéis ver en el diagrama, una LSTM tiene 3 puertas de este tipo, para proteger y controlar la cina transportadora. Podemos entender rápidamente su funcionamiento ayudándonos de este blog: http://colah.github.io/posts/2015-08-Understanding-LSTMs/

Y este blog también es muy interesante, para vosotros, curiosos: http://karpathy.github.io/2015/05/21/rnn-effectiveness/

### 4.1. Clasificación de imágenes con MNIST
Vamos a ver si podemos clasificar las imágenes del MNIST con una LSTM, y con qué precisión:

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import initializers

# Hyper parameters
batch_size = 128
nb_epoch = 10

# Parameters for MNIST dataset
img_rows, img_cols = 28, 28
nb_classes = 10

# Parameters for LSTM network
nb_lstm_outputs = 30
nb_time_steps = img_rows
dim_input_vector = img_cols

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('X_train original shape:', X_train.shape)
input_shape = (nb_time_steps, dim_input_vector)

X_train = X_train.astype('float32') / 255.
X_test = X_test.astype('float32') / 255.

Y_train = to_categorical(y_train, nb_classes)
Y_test = to_categorical(y_test, nb_classes)

print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```
# Construimos la LSTM
model = Sequential()
model.add(LSTM(nb_lstm_outputs, input_shape=input_shape))
model.add(Dense(nb_classes, activation="softmax"))
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```
# Entrenamos
model.fit(X_train, Y_train, epochs=nb_epoch, batch_size=batch_size, shuffle=True, validation_data=(X_test, Y_test), verbose=1)
```

```
# Evaluamos
evaluation = model.evaluate(X_test, Y_test, batch_size=batch_size, verbose=1)
print('Summary: Loss over the test dataset: %.2f, Accuracy: %.2f' % (evaluation[0], evaluation[1]))
```

### 4.2. Predicción de letras consecutivas en el alfabeto

```
# Naive LSTM to learn one-char to one-char mapping (https://machinelearningmastery.com/understanding-stateful-lstm-recurrent-neural-networks-python-keras/)
import numpy
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.utils import to_categorical

# fix random seed for reproducibility
numpy.random.seed(7)
# define the raw dataset
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
# create mapping of characters to integers (0-25) and the reverse
char_to_int = dict((c, i) for i, c in enumerate(alphabet))
int_to_char = dict((i, c) for i, c in enumerate(alphabet))

# prepare the dataset of input to output pairs encoded as integers
seq_length = 1
dataX = []
dataY = []
for i in range(0, len(alphabet) - seq_length, 1):
	seq_in = alphabet[i:i+seq_length]
	seq_out = alphabet[i + seq_length]
	dataX.append([char_to_int[char] for char in seq_in])
	dataY.append(char_to_int[seq_out])
	print(seq_in, '->', seq_out)

# reshape X to be [samples, time steps, features]
X = numpy.reshape(dataX, (len(dataX), seq_length, 1))

# normalize
X = X / float(len(alphabet))

# one hot encode the output variable
y = to_categorical(dataY)

# create and fit the model
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=1, verbose=2)

# summarize performance of the model
scores = model.evaluate(X, y, verbose=0)
print("Model Accuracy: %.2f%%" % (scores[1]*100))

# demonstrate some model predictions
for pattern in dataX:
	x = numpy.reshape(pattern, (1, len(pattern), 1))
	x = x / float(len(alphabet))
	prediction = model.predict(x, verbose=0)
	index = numpy.argmax(prediction)
	result = int_to_char[index]
	seq_in = [int_to_char[value] for value in pattern]
	print(seq_in, "->", result)
```

Ejemplo 1. Vamos a tratar de predecir el número viajeros de una aerolinea.

```
# LSTM for international airline passengers problem with regression framing
# https://machinelearningmastery.com/time-series-prediction-lstm-recurrent-neural-networks-python-keras/
!wget https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/airline/international-airline-passengers.csv

import numpy
import matplotlib.pyplot as plt
from pandas import read_csv
import math
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error

# convert an array of values into a dataset matrix
def create_dataset(dataset, look_back=1):
  dataX, dataY = [], []
  last_in_element = len(dataset) - (look_back + 1)
  for i in range(last_in_element):
    x = dataset[i:(i+look_back)]
    dataX.append(x)
    dataY.append(dataset[i+look_back])
  return numpy.array(dataX), numpy.array(dataY)

# fix random seed for reproducibility
numpy.random.seed(7)

# load the dataset
dataframe = read_csv('international-airline-passengers.csv', usecols=[1], engine='python', skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')
```

```
plt.plot(dataset)
```

```

# normalize the dataset
scaler = StandardScaler()
dataset = scaler.fit_transform(dataset).flatten()

# split into train and test sets
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[:-test_size], dataset[-test_size:]

# reshape into X=t and Y=t+1
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

# reshape input to be [samples, time steps, features]
trainX = numpy.expand_dims(trainX, -1)
testX = numpy.expand_dims(testX, -1)

# create and fit the LSTM network
model = Sequential()
model.add(LSTM(4, input_shape=(look_back, 1)))
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")
model.fit(trainX, trainY, epochs=20, batch_size=1, verbose=2)

# make predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

# invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

# calculate root mean squared error
trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))

# shift train predictions for plotting
# trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot = numpy.array([numpy.nan] * len(dataset))
trainPredictPlot[look_back:len(trainPredict)+look_back] = trainPredict.flatten()

# # shift test predictions for plotting
# testPredictPlot = numpy.empty_like(dataset)
testPredictPlot = numpy.array([numpy.nan] * len(dataset))
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1] = testPredict.flatten()

# plot baseline and predictions
plt.plot(scaler.inverse_transform(dataset.reshape(-1, 1)))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```



Ejemplo 2. Vamos a tratar de solucionar el problema del IMDB.

```
# imdb problem
# https://machinelearningmastery.com/sequence-classification-lstm-recurrent-neural-networks-python-keras/

# LSTM for sequence classification in the IMDB dataset
import numpy
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing import sequence

# fix random seed for reproducibility
SEED = 42
numpy.random.seed(SEED)

# load the dataset but only keep the top n words, zero the rest
top_words = 5000
word_index = imdb.get_word_index()
reverse_word_index = {word_index[w]: w for w in word_index}
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words, seed=SEED)

def decode(tokens, vocabulary) -> str:
  return " ".join(vocabulary.get(i-3, "[START]") for i in tokens)

print(f"Ejemplo de una review (raw): <{X_train[0]}>")
print(f"Ejemplo de una review (encoded-decoded): <{decode(X_train[0], reverse_word_index)}>")

# truncate and pad input sequences
max_review_length = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)

# create the model
embedding_vector_length = 32
model = Sequential()
model.add(Embedding(top_words, embedding_vector_length, input_length=max_review_length))
model.add(LSTM(128))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
print(model.summary())
model.fit(X_train, y_train, epochs=5, batch_size=32)

#  Final evaluation of the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

## Atención

En esta sección, nos adentraremos en una de las arquitecturas más revolucionarias en el campo del procesamiento del lenguaje natural (NLP) y el aprendizaje automático: los Transformers. Estas redes neuronales han marcado un hito significativo en la capacidad de modelar secuencias de datos de manera más efectiva, superando muchas de las limitaciones de las arquitecturas anteriores. En este módulo, exploraremos en detalle el funcionamiento interno de los Transformers, comenzando por uno de sus componentes fundamentales: la operación de atención.

La atención se inspira en el mecanismo de atención humano, que permite a los seres humanos enfocarse en partes específicas de una entrada, ignorando el resto. En el contexto de NLP, esto significa que el modelo puede asignar ponderaciones a diferentes partes de una oración, centrándose en las palabras más relevantes para la tarea en cuestión.

La consistencia de la operación de atención radica en su capacidad para calcular estas ponderaciones mediante una combinación lineal de valores de entrada, conocidos como consultas, claves y valores. Esta combinación ponderada permite al modelo enfocarse en las partes más importantes de la secuencia de entrada. La atención se calcula mediante una función de similitud, como el producto escalar o la función softmax, que determina qué elementos de la entrada son más relevantes para cada elemento de salida.

Es fundamental notar que la atención es una operación que tiene lugar de forma dinámica. Es decir, se calcula para cada muestra. No se aprende un esquema de atención general e igual para todas las muestras.

### Transformer

La arquitectura Transformer es un tipo de red neuronal diseñada para modelar relaciones a largo plazo en datos secuenciales, como texto o audio. Se destaca por su capacidad para capturar dependencias a larga distancia y su eficiencia en paralelización, lo que la convierte en una elección popular para tareas de procesamiento del lenguaje natural (NLP).

Especialmente relevante es su estructura encoder-decoder, conceptos herederos de la idea de "extracción de características" y "clasificador".

- **Positional Embedding**: La operación de atención es inmutable a permutaciones (no importa dónde estén los elementos en las secuencias de entrada). Para aportar dicha información de una manera aproximada al modelo, sumamos a la entrada del Transformer una señal de tipo sinusoidal que de alguna manera ayudan al modelo a entender que existe una relación de continuidad entre los elementos de la secuencia.

<img src="https://machinelearningmastery.com/wp-content/uploads/2022/01/PE5.png" alt="positionalEmb" border="0">


```
import tensorflow as tf
from tensorflow.keras.layers import Layer, ReLU, Dense, Embedding, LayerNormalization, Dropout
from tensorflow import matmul, cast, shape, reshape, transpose
from tensorflow.keras.activations import softmax
import numpy

class PositionalEncoding(Layer):
  def __init__(self, sequence_length, vocab_size, output_dim, **kwargs):
    super(PositionalEncoding, self).__init__(**kwargs)
    self.embeddings = Embedding(
      input_dim=vocab_size,
      output_dim=output_dim,
      trainable=False
    )
    self.positional_embeddings = Embedding(
      input_dim=sequence_length,
      output_dim=output_dim,
      trainable=False
    )

  def get_position_encoding(self, length, d, n=20000):
    P = numpy.zeros((length, d))
    for k in range(length):
      for i in numpy.arange(int(d/2)):
        denominator = numpy.power(n, 2*i/d)
        P[k, 2*i] = numpy.sin(k / denominator)
        P[k, 2*i + 1] = numpy.cos(k / denominator)
    return P

  def call(self, x):
    idxs = tf.range(tf.shape(x)[-1])
    embedded_x = self.embeddings(x)
    embedded_idxs = self.positional_embeddings(idxs)
    return embedded_x + embedded_idxs


vocab_size = 10000
seq_length = 10
embeddings_len = 512
batch_size = 128

positional_encodings = PositionalEncoding(seq_length, vocab_size, embeddings_len)
word_embeddings = numpy.random.randint(low=0, high=vocab_size, size=(batch_size, seq_length))
print("Positional Encoding output shape: ", positional_encodings(word_embeddings).shape)
```

- **Bloques de Atención (Attention Blocks)**: Los bloques de atención son la piedra angular de la arquitectura Transformer. Cada bloque de atención consiste en múltiples capas de atención, normalización y transformación. Dentro de un bloque de atención, se calculan las ponderaciones de atención entre todas las posiciones de la secuencia de entrada, permitiendo al modelo enfocarse en partes relevantes de la entrada en función del contexto. Este proceso de atención se realiza de manera paralela y eficiente, lo que permite capturar relaciones a largo plazo sin depender de la proximidad entre las palabras.

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fmachinelearningmastery.com%2Fwp-content%2Fuploads%2F2022%2F03%2Fdotproduct_1.png&f=1&nofb=1&ipt=c21477d77393cd848d6829c52cba5dd108e3d8c782c215e49239e11d46f06108&ipo=images">



```
import math

class DotProductAttention(Layer):
  def __init__(self, **kwargs):
    super(DotProductAttention, self).__init__(**kwargs)

  def call(self, queries, keys, values, d_k, mask=None):
    # asignar una puntuacion a las queries al compararlas con las keys, y escalado
    scores = matmul(queries, keys, transpose_b=True) / math.sqrt(cast(d_k, tf.float32))
    # enmascarado de las puntuaciones (si se ocultan palabras, o se precisa mantener la causalidad)
    if mask is not None:
      scores += -1e10 * mask # numero extremadamente pequeños
    # convertir puntuaciones a probabilidades (= pesos estadisticos)
    weigths = softmax(scores)
    return matmul(weigths, values)


batch_size = 128
d_k = 64
d_v = 64
seq_length = 10
queries = numpy.random.rand(batch_size, seq_length, d_k)
keys = numpy.random.rand(batch_size, seq_length, d_k)
values = numpy.random.rand(batch_size, seq_length, d_v)

attention = DotProductAttention()
print("DotProduct Attention: ", attention(queries, keys, values, d_k=d_k).shape)
```

```
class MultiHeadAttention(Layer):
  def __init__(self, h, d_k, d_v, d_model, **kwargs):
    super(MultiHeadAttention, self).__init__(**kwargs)
    self.attention = DotProductAttention()
    self.heads = h # numero de cabezas de atencion a usar
    self.d_k = d_k
    self.d_v = d_v
    self.W_q = Dense(d_k)
    self.W_k = Dense(d_k)
    self.W_v = Dense(d_v)
    self.W_o = Dense(d_model)

  def call(self, queries, keys, values, mask=None):
    # reordenar tensores para calcular todo en paralelo
    # print(queries.shape, keys.shape, values.shape)
    # input()
    q = self.reshaping(self.W_q(queries), self.heads, True)
    k = self.reshaping(self.W_k(keys), self.heads, True)
    v = self.reshaping(self.W_v(values), self.heads, True)

    # attention over these tensors
    o = self.attention(q, k, v, d_k=self.d_k, mask=mask)
    output = self.reshaping(o, self.heads, False)
    return self.W_o(output)

  def reshaping(self, x, heads, flag):
    if flag:
      x = reshape(x, shape=(shape(x)[0], shape(x)[1], heads, -1))
      x = transpose(x, perm=(0, 2, 1, 3))
      # dims = (BS, heads, seq_length, -1)
    else:
      x = transpose(x, perm=(0, 2, 1, 3))
      x = reshape(x, shape=(shape(x)[0], shape(x)[1], -1))
      # dims = (BS, seq_length, d_model)
    return x


nb_attention_heads = 8
d_model = 512
multihead_attn = MultiHeadAttention(nb_attention_heads, d_k, d_v, d_model)
print("Multihead Attention: ", multihead_attn(queries, keys, values).shape)
```

- **Capas Lineales**: Son capas feedforward completamente conectadas que siguen a los bloques de atención en la arquitectura Transformer. Estas capas realizan transformaciones lineales seguidas de funciones de activación no lineales, como la función de activación ReLU (Rectified Linear Unit). Estas transformaciones permiten al modelo capturar representaciones más complejas y abstractas de los datos de entrada, lo que facilita la extracción de características importantes para la tarea en cuestión.

```
class FeedForward(Layer):
  def __init__(self, d_ff: int = 2048, d_model: int = 512, **kwargs):
    super(FeedForward, self).__init__(**kwargs)
    self.fc1 = Dense(d_ff)
    self.fc2 = Dense(d_model)
    self.activation = ReLU()

  def call(self, x):
    z = self.fc1(x)
    return self.fc2(self.activation(z))
```


- **Conexiones Residuales y Normalización por Capas (Residual Connections and Layer Normalization)**: Para facilitar el entrenamiento de redes profundas, la arquitectura Transformer utiliza conexiones residuales entre las capas y normalización por capas después de cada capa de atención y transformación. Las conexiones residuales permiten que la información fluya más fácilmente a través de la red, evitando problemas de desvanecimiento del gradiente. La normalización por capas ayuda a estabilizar el proceso de entrenamiento al normalizar las activaciones en cada capa.

```
class AddNormalization(Layer):
  def __init__(self, **kwargs):
    super(AddNormalization, self).__init__(**kwargs)
    self.layer_norm = LayerNormalization()

  def call(self, x, residual):
    return self.layer_norm(x + residual)
```

En conjunto, estos elementos conforman una capa de procesamiento del encoder:

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Ftse1.mm.bing.net%2Fth%3Fid%3DOIP.ANp02TCmA6Nw0obCRoBFawAAAA%26pid%3DApi&f=1&ipt=38b1acac5bf7b27c9e7e09ce3004c6a4d26720333f544aee8e066e014fe1c77e&ipo=images" alt="EncoderTransformer" border="0">

```
class EncoderLayer(Layer):
  def __init__(self, h, d_k, d_v, d_model, d_ff, rate, **kwargs):
    super(EncoderLayer, self).__init__(**kwargs)
    self.multihead_attn = MultiHeadAttention(h, d_k, d_v, d_model)
    self.dropout = Dropout(rate)
    self.add_norm1 = AddNormalization()
    self.fc = FeedForward(d_ff, d_model)
    self.add_norm2 = AddNormalization()

  def call(self, x, padding_mask, training=False):
    # calcular atenciones y aplicar dropout
    z = self.dropout(self.multihead_attn(x, x, x, mask=padding_mask))
    z_norm = self.add_norm1(x, z)
    z = self.dropout(self.fc(z_norm), training=training)
    return self.add_norm2(z_norm, z)


class Encoder(Layer):
  def __init__(self, vocab_size, sequence_length, h, d_k, d_v, d_model, d_ff, nb_layers, rate, **kwargs):
    super(Encoder, self).__init__(**kwargs)
    self.positional_encoding = PositionalEncoding(sequence_length, vocab_size, d_model)
    self.dropout = Dropout(rate)
    self.encoder_layers = [
        EncoderLayer(h, d_k, d_v, d_model, d_ff, rate)
      for layer in range(nb_layers)
    ]

  def call(self, x, padding_mask, training):
    # generamos embeddings de la secuencia de entrada, y añadimos la pseudoinformacion temporal
    positional_encodings = self.positional_encoding(x)
    z = self.dropout(positional_encodings, training=training)
    # procesamos capa a capa el input, cada vez mayor nivel de abstraccion
    for i, layer in enumerate(self.encoder_layers):
      z = layer(z, padding_mask=padding_mask, training=training)
    return z


d_ff = 2048
dropout_rate = 0.25
nb_encoder_layers = 6
nb_attention_heads = 8
d_model = 512
batch_size = 128
d_k = 64
d_v = 64
seq_length = 10
vocab_size = 25
seq_length = 10
embeddings_len = 512

tokens = numpy.random.randint(0, vocab_size, size=(batch_size, seq_length))
encoder = Encoder(
  vocab_size=vocab_size,
  sequence_length=seq_length,
  h=nb_attention_heads,
  d_k=d_k,
  d_v=d_v,
  d_model=d_model,
  d_ff=d_ff,
  nb_layers=nb_encoder_layers,
  rate=dropout_rate
)

print("Input al Transfomer (secuencia de tokens): ", tokens)
print("Input al Transfomer (secuencia de tokens) shape: ", tokens.shape)
print("Encoder output: ", encoder(x=tokens, padding_mask=None, training=False).shape)
```
